```python
import pandas as pd
```

```python
inventory_parts = pd.read_csv('data/inventory_parts.csv')
parts = pd.read_csv('data/parts.csv')
colors = pd.read_csv('data/colors.csv')
```

```python
lego_parts = pd.merge(inventory_parts, parts,'left', on='part_num' )
```

```python
lego_parts = pd.merge(lego_parts, colors,'left', left_on='color_id', right_on='id')
lego_parts.head()
```

```python
lego_parts = lego_parts[["part_num", "name_x", "name_y", "img_url"]]
print(len(lego_parts))
print(len(lego_parts))
lego_parts.rename(columns={"part_num": "part_id", "name_x": 'part_name', 'name_y':'color_name', 'img_url': 'part_png'}, inplace=True)
```

```python
lego_parts.drop_duplicates(subset=['part_id', 'color_name', 'part_png'])
print(len(lego_parts))
lego_parts.head()
lego_parts.to_csv('data/lego_parts.csv')
```
